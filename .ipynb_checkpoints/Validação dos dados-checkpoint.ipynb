```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados_adm = pd.read_csv('SERVIDORES_ADM_15_07_2025.csv')
dados_jud = pd.read_csv('SERVIDORES_JUD_15_07_2025.csv')
dados_adm = dados_adm.drop(columns=['LOTA_COD_LOTACAO'])
#dados['LOTACAO_PAI'] = dados['LOTACAO_PAI'].replace('3¬™ TURMA RECURSAL/JEF/CE', '3¬™ TURMA RECURSAL')
dados = pd.concat([dados_adm, dados_jud], ignore_index=True)
dados.head()
```

```python
def buscar_matriculas(excel_path, lista_matriculas):
    xls_local = pd.ExcelFile(excel_path)

    for matricula in lista_matriculas:
        abas_encontradas = []

        for sheet_name in xls_local.sheet_names:
            df = xls_local.parse(sheet_name, dtype=str)
            if df.isin([matricula]).any().any():
                abas_encontradas.append(sheet_name)
        if abas_encontradas:
            abas_str = ', '.join(abas_encontradas)
            print(f"‚úÖ Matr√≠cula {matricula} encontrada na(s) aba(s): {abas_str}")
        else:
            print(f"‚ùå Matr√≠cula {matricula} n√£o encontrada em nenhuma aba.")


# Lista de LOTACAO_PAI para verificar
lotacoes = ['SECRETARIA ADMINISTRATIVA','DIRETORIA DO FORO','NUCLEO DE GESTAO DE PESSOAS','NUCLEO DE TECNOLOGIA DA INFORMA√áAO E COMUNICA√áAO',
            'NUCLEO DE INFRAESTRUTURA E ADMINISTRA√áAO PREDIAL','NUCLEO DE ESTRATEGIA, GOVERNAN√áA E INTEGRIDADE','NUCLEO DE AUDITORIA INTERNA',
            '1¬™ TURMA RECURSAL','1¬™ VARA - FORTALEZA-CE','2¬™ TURMA RECURSAL','2¬™ TURMA RECURSAL/JEF/CE','2¬™ VARA - FORTALEZA-CE',
            '3¬™ TURMA RECURSAL','3¬™ TURMA RECURSAL/JEF/CE','3¬™ VARA - FORTALEZA-CE',
            '4¬™ VARA - FORTALEZA-CE','5¬™ VARA - FORTALEZA-CE','6¬™ VARA - FORTALEZA-CE','7¬™ VARA - FORTALEZA-CE','8¬™ VARA - FORTALEZA-CE',
            '9¬™ VARA - FORTALEZA-CE','10¬™ VARA - FORTALEZA-CE','11¬™ VARA - FORTALEZA-CE','12¬™ VARA - FORTALEZA-CE','13¬™ VARA - JEF - FORTALEZA-CE',
            '14¬™ VARA - JEF - FORTALEZA-CE','15¬™ VARA - LIMOEIRO DO NORTE-CE','16¬™ VARA - JUAZEIRO DO NORTE-CE',
            '17¬™ VARA - JEF - JUAZEIRO DO NORTE-CE','18¬™ VARA - SOBRAL-CE','19¬™ VARA - JEF - SOBRAL-CE','20¬™ VARA - FORTALEZA-CE',
            '21¬™ VARA - JEF - FORTALEZA-CE','22¬™ VARA - CRATE√öS-CE','23¬™ VARA - QUIXAD√Å-CE','24¬™ VARA - TAU√Å-CE','25¬™ VARA - IGUATU-CE',
            '26¬™ VARA - JEF - FORTALEZA - CE','27¬™ VARA- ITAPIPOCA-CE','28¬™ VARA - JEF - FORTALEZA-CE','29¬™ VARA - JEF - LIMOEIRO DO NORTE - CE',
            '30¬™ VARA - JEF - JUAZEIRO DO NORTE - CE','31¬™ VARA - JEF - SOBRAL - CE','32¬™ VARA - FORTALEZA-CE','33¬™ VARA - FORTALEZA-CE',
            '34¬™ VARA - MARACANA√ö-CE','35¬™ VARA - JEF - MARACANA√ö-CE']

# Aplicar a fun√ß√£o para cada filtro
for lotacao in lotacoes:
    print(f"\nüîé Verificando lota√ß√£o: {lotacao}")
    matriculas_lotacao = dados[dados['LOTACAO_PAI'] == lotacao]['MATRICULA'].dropna().unique()
    buscar_matriculas('JULHO.xlsx', matriculas_lotacao.tolist())
```
