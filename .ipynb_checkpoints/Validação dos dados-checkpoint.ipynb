```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados_adm = pd.read_csv('SERVIDORES_ADM_15_07_2025.csv')
dados_jud = pd.read_csv('SERVIDORES_JUD_15_07_2025.csv')
dados_adm = dados_adm.drop(columns=['LOTA_COD_LOTACAO'])
#dados['LOTACAO_PAI'] = dados['LOTACAO_PAI'].replace('3Âª TURMA RECURSAL/JEF/CE', '3Âª TURMA RECURSAL')
dados = pd.concat([dados_adm, dados_jud], ignore_index=True)
dados.head()
```

```python
def buscar_matriculas(excel_path, lista_matriculas):
    xls_local = pd.ExcelFile(excel_path)

    for matricula in lista_matriculas:
        abas_encontradas = []

        for sheet_name in xls_local.sheet_names:
            df = xls_local.parse(sheet_name, dtype=str)
            if df.isin([matricula]).any().any():
                abas_encontradas.append(sheet_name)
        if abas_encontradas:
            abas_str = ', '.join(abas_encontradas)
            print(f"âœ… MatrÃ­cula {matricula} encontrada na(s) aba(s): {abas_str}")
        else:
            print(f"âŒ MatrÃ­cula {matricula} nÃ£o encontrada em nenhuma aba.")


# Lista de LOTACAO_PAI para verificar
lotacoes = ['SECRETARIA ADMINISTRATIVA','DIRETORIA DO FORO','NUCLEO DE GESTAO DE PESSOAS','NUCLEO DE TECNOLOGIA DA INFORMAÃ‡AO E COMUNICAÃ‡AO',
            'NUCLEO DE INFRAESTRUTURA E ADMINISTRAÃ‡AO PREDIAL','NUCLEO DE ESTRATEGIA, GOVERNANÃ‡A E INTEGRIDADE','NUCLEO DE AUDITORIA INTERNA',
            '1Âª TURMA RECURSAL','1Âª VARA - FORTALEZA-CE','2Âª TURMA RECURSAL','2Âª TURMA RECURSAL/JEF/CE','2Âª VARA - FORTALEZA-CE',
            '3Âª TURMA RECURSAL','3Âª TURMA RECURSAL/JEF/CE','3Âª VARA - FORTALEZA-CE',
            '4Âª VARA - FORTALEZA-CE','5Âª VARA - FORTALEZA-CE','6Âª VARA - FORTALEZA-CE','7Âª VARA - FORTALEZA-CE','8Âª VARA - FORTALEZA-CE',
            '9Âª VARA - FORTALEZA-CE','10Âª VARA - FORTALEZA-CE','11Âª VARA - FORTALEZA-CE','12Âª VARA - FORTALEZA-CE','13Âª VARA - JEF - FORTALEZA-CE',
            '14Âª VARA - JEF - FORTALEZA-CE','15Âª VARA - LIMOEIRO DO NORTE-CE','16Âª VARA - JUAZEIRO DO NORTE-CE',
            '17Âª VARA - JEF - JUAZEIRO DO NORTE-CE','18Âª VARA - SOBRAL-CE','19Âª VARA - JEF - SOBRAL-CE','20Âª VARA - FORTALEZA-CE',
            '21Âª VARA - JEF - FORTALEZA-CE','22Âª VARA - CRATEÃšS-CE','23Âª VARA - QUIXADÃ-CE','24Âª VARA - TAUÃ-CE','25Âª VARA - IGUATU-CE',
            '26Âª VARA - JEF - FORTALEZA - CE','27Âª VARA- ITAPIPOCA-CE','28Âª VARA - JEF - FORTALEZA-CE','29Âª VARA - JEF - LIMOEIRO DO NORTE - CE',
            '30Âª VARA - JEF - JUAZEIRO DO NORTE - CE','31Âª VARA - JEF - SOBRAL - CE','32Âª VARA - FORTALEZA-CE','33Âª VARA - FORTALEZA-CE',
            '34Âª VARA - MARACANAÃš-CE','35Âª VARA - JEF - MARACANAÃš-CE']

# Aplicar a funÃ§Ã£o para cada filtro
for lotacao in lotacoes:
    print(f"\nğŸ” Verificando lotaÃ§Ã£o: {lotacao}")
    matriculas_lotacao = dados[dados['LOTACAO_PAI'] == lotacao]['MATRICULA'].dropna().unique()
    buscar_matriculas('JULHO.xlsx', matriculas_lotacao.tolist())
```
