```python
import requests
import xml.etree.ElementTree as ET

# URL do webservice
url = "https://webservice-h.jfce.jus.br/sarh_new/buscarFuncionarios"

# Fazendo a requisição
response = requests.get(url)

# Verificando se a requisição foi bem-sucedida
if response.status_code == 200:
    xml_content = response.content
    root = ET.fromstring(xml_content)

    # Iterando sobre os elementos <funcionario>
    for funcionario in root.findall(".//funcionario"):
        nome = funcionario.findtext("nome")
        matricula = funcionario.findtext("matricula")
        email = funcionario.findtext("email")
        cargo = funcionario.findtext("cargo")
        lotacao = funcionario.findtext("lotacao")
        nascimento = funcionario.findtext("nascimento")

        print(f"Nome: {nome}")
        print(f"Matrícula: {matricula}")
        print(f"E-mail: {email}")
        print(f"Cargo: {cargo}")
        print(f"Lotação: {lotacao}")
        print(f"Nascimento: {nascimento}")
        print("-" * 40)
else:
    print(f"Erro ao acessar o webservice: {response.status_code}")
```

```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd

url = "https://webservice-h.jfce.jus.br/sarh_new/buscarFuncionarios"
response = requests.get(url)
dados_funcionarios = []

if response.status_code == 200:
    xml_content = response.content
    root = ET.fromstring(xml_content)
    for funcionario in root.findall(".//funcionario"):
        dados_funcionarios.append({
            "id": funcionario.findtext("id"),
            "matricula": funcionario.findtext("matricula"),
            "nome": funcionario.findtext("nome"),
            "email": funcionario.findtext("email"),
            "login": funcionario.findtext("login"),
            "cpf": funcionario.findtext("cpf"),
            "exercicio": funcionario.findtext("exercicio"),
            "classe": funcionario.findtext("classe"),
            "padrao": funcionario.findtext("padrao"),
            "lotacao": funcionario.findtext("lotacao"),
            "siglaLotacao": funcionario.findtext("siglaLotacao"),
            "nascimento": funcionario.findtext("nascimento"),
            "cargo": funcionario.findtext("cargo")
        })
        
        # Acessa o último funcionário inserido
        ultimo = dados_funcionarios[-1]
        print(f"Nome: {ultimo['nome']}")
        print(f"Matrícula: {ultimo['matricula']}")
        print(f"E-mail: {ultimo['email']}")
        print(f"Cargo: {ultimo['cargo']}")
        print(f"Lotação: {ultimo['lotacao']}")
        print(f"Nascimento: {ultimo['nascimento']}")
        print("-" * 40)

    # Criar DataFrame
    df = pd.DataFrame(dados_funcionarios)
else:
    print(f"Erro ao acessar o webservice: {response.status_code}")
```

```python
# Converter nascimento para datetime
df["nascimento"] = pd.to_datetime(df["nascimento"], errors="coerce")

# Calcular idade com base na data atual
df["idade"] = df["nascimento"].apply(lambda x: (pd.to_datetime("today") - x).days // 365 if pd.notnull(x) else None)
```

```python
# Salvar como CSV
df.to_csv("funcionarios_jfce.csv", index=False)

# Salvar como Excel
df.to_excel("funcionarios_jfce.xlsx", index=False)
```
